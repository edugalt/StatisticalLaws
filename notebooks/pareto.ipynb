```python
datas=["Prussia1886","Saxony1886","ItalianCities","GB","Oldenburg","Ireland","Peru","Augsburg1471"]#"Basel","Basel1454","Paris",,"Florence"
```

```python
xscale("log")
yscale("log")
markers = ["o","v","s","p","*","x","+",">","<","X","p","o"]
i=0
for d in datas:
    data=pd.read_csv("../data/income/pareto-"+d+".csv",names=["N","x"])
    plot(data['N'],data['x'],marker=markers[i],label=d)
    i+=1
xlabel("x: income")
ylabel("N: population with income > x")
#xlim([0.15,2*10**5])
#legend(loc=0)
xlim([2,2*10**6])
legend(loc=0)
title("Income distribution (Pareto data)")
savefig("fig/pareto.png",dpi=300,bbox_inches='tight')
show()
```
