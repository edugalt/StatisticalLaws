Note on data:

Citations retrieved from Google Scholar on June 7, 2024

The series for the 3 XXI century papers were obtained from the author's profile, year by year
The series for the 2 XX century papers were obtained from the "cited-by" list, restricting to each year (the value is the reported estimated one).

```python
yzipf=[1949,1950,1955,1960,1965,1970,1975,1980,1985,1990,1995,2000]
for i in range(23):
    yzipf.append(2000+i+1)
czipf =[2,9,8,18,30,37,54,47,49,94,139,235,322,364,407,492,566,589,636,637,627,555,638,717,737,846,742,796,703,696,696,728,829,691,670]
```

```python
ykleiber=[1935,1940,1945,1948,1949,1950,1955,1960,1965,1970,1975,1980,1985,1990,1995,2000]
for i in range(23):
    ykleiber.append(2000+i+1)
ckleiber =[0,0,0,8,6,12,9,11,14,15,16,18,23,20,21,16,22,15,27,20,31,28,17,28,38,39,62,87,68,100,79,92,93,94,102,105,111,104,96]
```

```python
ybarabasi=[]
for i in range(23):
    ybarabasi.append(2000+i+1)
cbarabasi = [184,327,524,795,1029,1442,1777,1899,2242,2390,2424,2653,2743,2821,2751,2729,2591,2438,2461,2293,2205,2076,2030]
```

```python
yclauset =[]
for i in range(16):
    yclauset.append(2008+i)
cclauset=[110,208,346,472,626,769,890,882,917,872,863,895,844,861,737,667]
```

```python
ybettencourt=[]
for i in range(17):
    ybettencourt.append(2007+i)
cbettencourt=[11,30,43,65,78,113,129,203,211,215,239,226,256,266,318,282,220]
```

```python
plot(ykleiber,ckleiber,"-s",lw=3,label="Kleiber 1932")
plot(yzipf,czipf,"-o",label="Zipf 1949")
plot(ybarabasi,cbarabasi,"-^",label="Barabsi-Albert 1999")
plot(ybettencourt,cbettencourt,"-v",label="Bettencourt et al. 2007")
plot(yclauset,cclauset,"-*",label="Clauset et al. 2009")
yscale("log")
xlim(1945,2024)
ylim(6,4000)
xlabel("Year")
ylabel("Number of citations in a year")
legend()
savefig("fig/bibliometric.png",dpi=300,bbox_inches='tight')
show()
```
